**NLTK: Sentiment Samples**

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
def sentiment_value(paragraph):
    analyser = SentimentIntensityAnalyzer()
    result = analyser.polarity_scores(paragraph)
    score = result['compound']
    return round(score,1)
```

```python
import nltk
nltk.downloader.download('vader_lexicon')
```

```python
sentiment_1 = AmzRv['Reviews'][99314]
print(sentiment_1)
print('Sentiment: ')
print(sentiment_value(sentiment_1))
```

```python
sentiment_2 = AmzRv['Reviews'][1231]
print(sentiment_2)
print('Sentiment: ')
print(sentiment_value(sentiment_2))
```

```python
sentiment_3 = AmzRv['Reviews'][9001]
print(sentiment_3)
print('Sentiment: ')
print(sentiment_value(sentiment_3))
```

**Assigning Sentiment value for 10000 reviews**

```python
for i in range(0,10000):
    all_sent_values.append(sentiment_value(all_reviews[i])) 
```

```python
len(all_sent_values)
```

```python
temp_data = AmzRv[0:10000]
```

```python
temp_data.shape
```

**Assigning Sentiment against Sentiment Values**

**Intervals**

-1, -0.5 : 1, V.Negative

-0.5, 0 : 2,  Negative

 0 : 3,       Neutral

 0, 0.5 : 4,  Positive
 
 0.5, 1 : 5,  V.Positive

```python
SENTIMENT_VALUE = []
SENTIMENT = []

for i in range(0,10000):
    sent = all_sent_values[i]

    if (sent<=1 and sent>=0.5):
        SENTIMENT.append('V.Positive')
        SENTIMENT_VALUE.append(5)
    
    elif (sent<0.5 and sent>0):
        SENTIMENT.append('Positive')
        SENTIMENT_VALUE.append(4)
    
    elif (sent==0):
        SENTIMENT.append('Neutral')
        SENTIMENT_VALUE.append(3)
    
    elif (sent<0 and sent>=-0.5):
        SENTIMENT.append('Negative')
        SENTIMENT_VALUE.append(2)
    
    else:
        SENTIMENT.append('V.Negative')
        SENTIMENT_VALUE.append(1)
```

```python
temp_data['SENTIMENT_VALUE'] = SENTIMENT_VALUE
temp_data['SENTIMENT'] = SENTIMENT
```

**Matching Ratings with Sentiment Value and checking Accuracy of this Match**

```python
# Sentiment Values which is different by more than 1 from Rating value
counter = 0
for i in range(0,10000):
    if (abs(temp_data['Rating'][i]-temp_data['SENTIMENT_VALUE'][i])>1):
        counter += 1
```

```python
counter
```

```python
accuracy = (temp_data.shape[0]-counter)/temp_data.shape[0]
```

```python
percent_accuracy = accuracy*100
```

```python
percent_accuracy
```

**Thus, 75.94 % equal values of Rating and Sentiment Values (+/- 1)**
